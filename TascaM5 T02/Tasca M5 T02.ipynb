```python
import pandas as pd        
import numpy as np
import seaborn as sns   # de la columna RM del dataset  faig una columna dummy que em mostra si la casa te més de 6 hab
                        #posa un 1 i un 0 si té menys habitacions


huston['RM_DUMMY'] = np.where(huston['RM']>6, 1, 0) 
rm_dummy = huston[['RM_DUMMY']].copy()

z_scaler = preprocessing.StandardScaler()
iris_z = z_scaler.fit_transform(rm_dummy)
iris_z = pd.DataFrame(iris_z)



rm_dummy.describe()
```

```python
sns.jointplot(data=iris_z, color='m', x=0, y=1);
```
